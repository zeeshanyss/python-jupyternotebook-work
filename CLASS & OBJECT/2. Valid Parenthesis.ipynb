```python
def isBalanced(s):
    if not s:
        return ""
    stack = []
    for char in s:
        if char in '{[(':
            stack.append(char)
        elif char == ')':
            if (not stack or stack[-1] != ')'):
                return False
            stack.pop()
        elif char == ']':
            if (not stack or stack[-1] != ']'):
                return False
            stack.pop()
        elif char == '}':
            if (not stack or stack[-1] != '}'):
                return False
            stack.pop()
    if not stack:
        return True
    return False


s = input()
if isBalanced(s):
    print("True")
else:
    print("False")
```
