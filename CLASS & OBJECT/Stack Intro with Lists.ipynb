```python
# Stack : Last In First Out
# Properties:
# 1. POP : removing Top/Last element - using pop() function
# 2. PUSH : inserting element last - using append() function
# 3. TOP : get the top/last ekement - l[-1] [len(l) - 1]
# 4. SIZE : total elements - (len(list))
# 5. isEMpty : Stack - if len(s) == 0?
```

```python
stack = []
stack.append(10)
stack.append(20)
stack.append(30)
print(stack)
```

```python
# POP

stack.pop()
```

```python
# PUSH

stack.append(40)
stack
```

```python
# TOP
stack[-1]
```

```python
# isEmpty

len(stack) == 0
```

```python
# SIZE

len(stack)
```

```python
 # INBUILT STACK: LIFO

import queue
s = queue.LifoQueue()
s.put(1)
s.put(5)
s.put(15)

while not s.empty():
    print(s.get())
```

```python
 # INBUILT QUEUE: FIFO

import queue
s = queue.Queue()
s.put(1)
s.put(5)
s.put(15)

while not s.empty():
    print(s.get())
```
