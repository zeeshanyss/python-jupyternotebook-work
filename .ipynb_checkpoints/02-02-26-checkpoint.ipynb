```python
arr = [1, 0, 1, 0, 0, 1]

start = 0
end = len(arr) - 1

while start < end:

    if arr[start] == 0:
        start += 1
    else:
        arr[start], arr[end] = arr[end], arr[start]
        end -= 1

print(arr)
```

```python
arr = [2, 0, 1, 2, 1, 0]

start = 0
mid = 0
end = len(arr) - 1

while mid <= end:

    if arr[mid] == 0:
        arr[start], arr[mid] = arr[mid], arr[start]
        start += 1
        mid += 1

    elif arr[mid] == 1:
        mid += 1

    else:  
        arr[mid], arr[end] = arr[end], arr[mid]
        end -= 1

print(arr)
```
