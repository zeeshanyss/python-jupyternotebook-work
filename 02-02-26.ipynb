```python
arr = [1, 0, 1, 0, 0, 1]

start = 0
end = len(arr) - 1

while start < end:

    if arr[start] == 0:
        start += 1
    else:
        arr[start], arr[end] = arr[end], arr[start]
        end -= 1

print(arr)
```

```python
# WAP to print the sum of first n natural numbers
def sumn(n):
    if n == 1:  # BASE CASE
        return 1

    smallOutput = sumn(n - 1)  # Induction Hypothesis
    return smallOutput + n     # Induction Step

print(sumn(5))
```

```python
def sumlist(arr):
    if len(arr) == 0:        # BASE CASE
        return 0

    # Induction Hypothesis
    smallOutput = sumlist(arr[1:])

    # Induction Step
    return arr[0] + smallOutput

arr = [1, 2, 3, 4, 5]
print(sumlist(arr))
```
