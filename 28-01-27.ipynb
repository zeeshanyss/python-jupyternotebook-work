```python
n = int(input())

i = 1
while i <= n:

 
    j = 1
    while j <= n - i:
        print(" ", end="")
        j += 1

    k = 1
    while k <= i:
        print((2*i - 2) + k, end="")
        k += 1

 
    l = 2*i - 2
    while l >= i:
        print(l, end="")
        l -= 1

    print()
    i += 1
```

```python
# Program to print the Fibonacci series till n terms

n = int(input())

num1 = 0
num2 = 1

print(num1, end=" ")
print(num2, end=" ")

for i in range(3, n + 1):
    s = num1 + num2
    print(s, end=" ")

    # SWAP
    num1 = num2
    num2 = s
```

```python
# WAP to count the occurence of character in the string

1. declare empty dictionary
2. traverse the string
3. check if the character is present inside the dictionary
4. if present - increase the count, otherwise create the emtry of the element strating from 1 \



```

```python
s = input()

d = {}
for char in s:
    if char in d:
        d[char] += 1
    else:
        d[char] = 1
        
print("Character Occurrence:")
print(d)
```
