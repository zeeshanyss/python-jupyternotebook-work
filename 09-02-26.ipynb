```python
# Replace 'pi' with 3.14 ; abcpi - abc3.14

def replacepi(s):
    #base case
    if len(s) == 0:
        return s
    #smallOutput = replacepi(s[1:0])
    if s[0] == "p" and s[1] == "i":
        return "3.14" + replacepi(s[2:])
    else:
        return s[0] + replacepi(s[1:])

replacepi("abcpi")
```
