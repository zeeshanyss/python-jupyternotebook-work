```python
a = int(input())
b = int(input())

for i in range(min(a, b), 0, -1):
    if a % i == 0 and b % i == 0:
        gcd = i
        break

print("GCD =", gcd)
```

```python
a = int(input())
b = int(input())

i = max(a, b)

while True:
    if i % a != 0 or i % b != 0:
        i += 1
        continue

    lcm = i
    break

print("LCM =", lcm)
```
