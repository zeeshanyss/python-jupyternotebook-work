```python
n = int(input())

i = 0
while i < n:
    j = 0
    while j < i:
        print(" ", end="")
        j += 1

    k = 0
    while k < n - i:
        print("*", end="")
        k += 1

    print()
    i += 1

i = n - 2
while i >= 0:
    j = 0
    while j < i:
        print(" ", end="")
        j += 1

    k = 0
    while k < n - i:
        print("*", end="")
        k += 1

    print()
    i -= 1
```

```python
#reverse the element of a liist - using Swapping

n = [1, 2, 3, 4, 5]

i = 0
j = len(n) - 1

while i < j:
    n[i], n[j] = n[j], n[i]  
    i += 1
    j -= 1

print(n)
```

```python
n = int(input())

i = 1
while i <= n:

    j = 1
    while j <= n - i:
        print(" ", end="")
        j += 1

    k = 0
    while k < i:
        print(i + k, end="")
        k += 1

    l = 1
    while l < i:
        print(i + i - l - 1, end="")
        l += 1

    print()
    i += 1
```
