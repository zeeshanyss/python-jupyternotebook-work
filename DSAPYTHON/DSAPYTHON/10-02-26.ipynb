```python
# Binary Search - 

def binarySearch(arr, k):
    n = len(arr)
    si = 0
    ei = (n - 1)
    while (si <= ei):
        mid = (si + ei) // 2
        if arr[mid] == k :
            return mid
        elif arr[mid] < k :
            si = mid + 1
        else:
            ei = mid - 1
    return -1

binarySearch([1, 2, 3, 15, 19, 28, 45, 98], 3)
```

```python
def binarySearchBetter(arr, ele, si, ei):
    if (si > ei):
        return -1
    mid = (si + ei) // 2
    if(ele == arr[mid]):
        return mid
    elif (ele < arr[mid]):
        return binarySearchBetter(arr, ele , si, mid - 1)
    else:
        return binarySearchBetter(arr, ele, ei, mid + 1)

binarySearchBetter([1, 2, 4, 6, 9, 46, 56, 102], 3, 0, 8)
    
```

```python
def partition(a, si, ei):
    pivot = a[si]

    # Count elements smaller than pivot
    c = 0
    for i in range(si, ei + 1):
        if a[i] < pivot:
            c += 1

    # Place pivot at correct position
    a[si + c], a[si] = a[si], a[si + c]
    pivot_index = si + c

    # Rearrange elements around pivot
    while si < ei:
        if a[si] < pivot:
            si += 1
        elif a[ei] >= pivot:
            ei -= 1
        else:
            a[si], a[ei] = a[ei], a[si]
            si += 1
            ei -= 1

        return pivot_index
        
partition([1, 2, 4, 6, 9, 46, 56, 102], 3, 7)
```

```python
def tower_of_hanoi(n, a, b,c):
    #a = source, b = auxiliary, c = destination

    # Base Case
    if n == 1:
        print("Move 1st disk from", a, "to", c)
        return

    # Move n-1 disks from source to auxiliary
    tower_of_hanoi(n-1, a, c, b)

    # Move nth disk to destination
    print("Move", n, "th disk from", a, "to", c)

    # Move n-1 disks from auxilliary to destination
    tower_of_hanoi(n-1, b, a,c)

tower_of_hanoi(3, "source", "aux", "dest")
```
