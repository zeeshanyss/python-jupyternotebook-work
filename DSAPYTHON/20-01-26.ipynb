```python
n = int(input())

first = int(input())
largest = first
second = first

for i in range(n - 1):
    x = int(input())

    if x > largest:
        second = largest
        largest = x

    if x < largest and x > second:
        second = x

print(second)
```

```python
n = int(input())

arr = [0] * n

i = 0
while i < n:
    arr[i] = int(input())
    i = i + 1

k = int(input())

i = 0
while i < k:
    first = arr[0]

    j = 0
    while j < n - 1:
        arr[j] = arr[j + 1]
        j = j + 1

    arr[n - 1] = first
    i = i + 1

print(arr)
```

```python
temp = float(input())

if temp >= 85.0 and temp <= 91.0:
    print("Serious Hypothermia")
if temp > 91.0 and temp < 95.0:
    print("Mild Hypothermia")
if temp >= 95.0 and temp <= 98.0:
    print("Normal Temperature")
if temp > 98.0 and temp <= 100.0:
    print("Mild Fever")
if temp > 100.0 and temp <= 105.0:
    print("High Fever")
```
