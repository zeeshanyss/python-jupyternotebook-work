```python
# Replace 'pi' with 3.14 ; abcpi - abc3.14

def replacepi(s):
    #base case
    if len(s) == 0:
        return s
    #smallOutput = replacepi(s[1:0])
    if s[0] == "p" and s[1] == "i":
        return "3.14" + replacepi(s[2:])
    else:
        return s[0] + replacepi(s[1:])

replacepi("abcpi")
```

```python
# Binary Search Recursive

def binaryRecursive(arr, k, si, ei):
    if si > ei:
        return -1

    mid = (si + ei) // 2

    if arr[mid] == k:
        return mid
    elif arr[mid] < k:
        return binaryRecursive(arr, k, mid + 1, ei)
    else:
        return binaryRecursive(arr, k, si, mid - 1)


ans = binaryRecursive([1, 3, 4, 5, 7, 8, 11, 15, 16, 19, 23], 15, 0, 10)
print(ans)
```

```python
def binarySearch(arr, ele, si, ei):
    if si > ei:
        return -1
    
    mid = (si + ei) // 2
    
    if ele == arr[mid]:
        return mid
    elif ele < arr[mid]:
        return binarySearch(arr, ele, si, mid - 1)
    else:
        return binarySearch(arr, ele, mid + 1, ei)


ans = binarySearch([1, 3, 4, 5, 7, 8, 11, 15, 16, 19, 23], 15, 0, 10)
print(ans)
```
