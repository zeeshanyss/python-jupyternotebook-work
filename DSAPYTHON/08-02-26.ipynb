```python
# Kadane's Algo

def kadanesAlgo(arr): #Return max sum of subarray
    res = arr[0]
    maxSum = arr[0]

    for i in range(1, len(arr)):
        maxSum = max(maxSum + arr[i], arr[i])
        res = max(maxSum, res)
    return res

arr = [2, 3, -8, 7, -1, 2, 3]
print(kadanesAlgo(arr))
```

```python
def firstIndex(arr, x):
    # Base Case 
    if len(arr) == 0:
        return -1

    # Check for the 0th Index
    if arr[0] == x:
        return 0
    smallOutput = firstIndex(arr[1:], x)

    if smallOutput == -1:
        return -1
    else:
        return smallOutput + 1

a = [1, 2, 4 , 5 , 6 ,1, 8]
x = 4
print(firstIndex(a, x))
```

```python
# Return the first index of integer x present in the array (without using silicing)

def firstIndexBetter(arr, x, si):
    #base case
    if len(arr) == si:
        return -1

    if arr[si] == x:
        return si

    smallOutput = firstIndexBetter(arr, x, si + 1)
    return smallOutput

a = [1, 2, 4 , 5 , 6 ,1, 8]
x = 4
print(firstIndexBetter(a, x, 0))
```

```python
# Return the last iindex of integer x present in the array (use slicing)

def lastIndex(arr, x):
    if len(arr) == 0:
        return -1

    if arr[len(arr) - 1] == x:
        return len(arr) - 1

    smallOutput = lastIndex(arr[:len(arr) - 1], x)

    if smallOutput == -1:
        return -1
    return smallOutput
        
a = [1, 2, 4 , 5 , 6 ,1, 8]
x = 4
print(lastIndex(a, x))
```

```python
# return the last index of integer x present in the array (use index reference)

def lastIndexBetter(arr, x, si):
    if si == len(arr):
        return -1

    smallOutput = lastIndexBetter(arr, x, si + 1)

    if smallOutput != -1:
        return smallOutput
    else:
        if arr[si] == x:
            return si
        return -1


        
a = [1, 2, 4 , 5 , 6 ,1, 8]
x = 4
print(lastIndexBetter(a, x, 0))
```

```python
# Given a string, compute recursively a new string where all 'x' chars have been removed xaxbx  ab

def removex(s):
    #base case
    if len(s) == 0:
        return s

    # induction hypothesis

    smallOutput = removex(s[1:])

    #induction step
    if s[0] == 'x':
        return smallOutput
    else:
        return s[0] + smallOutput

removex("xaxbx")
```
