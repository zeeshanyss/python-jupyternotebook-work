```python
n = int(input())
temp = n
sum = 0

while n > 0:
    digit = n % 10
    sum = sum + digit * digit * digit
    n = n // 10

if sum == temp:
    print("Armstrong")
if sum != temp:
    print("Not Armstrong")
```
